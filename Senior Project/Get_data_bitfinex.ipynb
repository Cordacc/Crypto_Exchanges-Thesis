```python
import pandas as pd
import math
import os.path
import time
import bitfinex
from datetime import timedelta, datetime
```

```python
api_v2 = bitfinex.bitfinex_v2.api_v2()

def fetch_data_bitfinex(start, stop, symbol, interval, tick_limit, step, save = False):
    filename = '%s-%s-bitfinex.csv' % (symbol, interval)
    if os.path.isfile(filename): data_df = pd.read_csv(filename)
    # Create api instance
    api_v2 = bitfinex.bitfinex_v2.api_v2()
    data = []
    start = start - step
    while start < stop:
        start = start + step
        end = start + step
        res = api_v2.candles(symbol=symbol, interval=interval,
                             limit=tick_limit, start=start,
                             end=end)
        data.extend(res)
        time.sleep(2)
    names = ['time', 'open', 'close', 'high', 'low', 'volume']
    df = pd.DataFrame(data, columns=names)
    df.drop_duplicates(inplace=True)
    df['time'] = pd.to_datetime(df['time'], unit='ms')
    df.set_index('time', inplace=True)
    df.sort_index(inplace=True)
    if save > 0: df.to_csv(filename)
    return data

# Define query parameters
bin_size = '1m' # This will return minute data
limit = 1000    # We want the maximum of 1000 data points 
# Set step size
time_step = 60000000
# Define the start date 
t_start = datetime(2017, 12, 1, 0, 0)
t_start = time.mktime(t_start.timetuple()) * 1000
# Define the end date
t_stop = datetime(2019, 12, 1, 0, 0)
t_stop = time.mktime(t_stop.timetuple()) * 1000
```

```python
pair_data = fetch_data_bitfinex(start=t_start, stop=t_stop, symbol='btcusd',
                       interval=bin_size, tick_limit=limit, 
                       step=time_step, save = True)
```

```python
for i in ['btcusd','eosusd','ethusd','iotusd','ltcusd','neousd','xrpusd','ethbtc']:
    fetch_data_bitfinex(start=t_start, stop=t_stop, symbol=i,
                       interval=bin_size, tick_limit=limit, 
                       step=time_step, save = True)
```
